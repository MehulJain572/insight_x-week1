```python
#question 1
def sum(a,b):
    return a+b

num=int(input("enter first number"))
num2=int(input("enter second number"))
z=sum(num,num2)
print(z)
```

```python
#question 2
def greeting(name="guest"):
    str=f"welcome {name}"
    return str

num=str(input("please enter your name"))
if num=="":
    ans_num=greeting()
else:
    ans_num=greeting(num)

print(ans_num)
```
