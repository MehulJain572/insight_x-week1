```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#loading the dataset from seaborn's built-in library
df=sns.load_dataset('iris')
print(df.head())

#data cleaning
#1.) check for missing values
print(df.isnull().sum())

#2.) check for and remove duplicates
duplicate_count=df.duplicated().sum()
print(f"number of duplicates rows found:{duplicate_count}\n")
if duplicate_count>0:
    df=df.drop_duplicates()

#summary statistics
print(df.describe())

#visual analysis
sns.set_theme(style="whitegrid")
plt.figure(figsize=(15,10))

#visualization 1: species distribution (count plot)
plt.subplot(2,2,1)
sns.countplot(x='species', data=df, hue='species', palette='viridis', legend=False)
plt.title('distribution of iris species')

#visualization 2: feature relationships (pair plot)
sns.pairplot(df,hue='species',height=2.5)
plt.suptitle("pairwise relationships by species", y=1.02)

#visualization 3: coorelation heatmap
plt.figure(figsize=(8,6))
#only correlate numeric columns
numeric_df=df.drop(columns=['species'])
correlation_matrix=numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True,cmap='coolwarm',linewidth=0.5)
plt.title('feature correlation heatmap')

#all plots:
plt.show()
```
